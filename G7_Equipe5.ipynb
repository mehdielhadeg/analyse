```python
# we only want to work on tvshows
tv_series = df[df['type'] == 'TV Show']

# Extract the number of seasons from the 'duration' column
tv_series['num_seasons'] = tv_series['duration'].str.extract('(\d+)').astype(int)

# Count the number of unique seasons for each TV series
season_counts = tv_series.groupby('title')['num_seasons'].max()

# Sort the counts in descending order
top_5_series = season_counts.sort_values(ascending=False).head(5)


top_5_series
```

```python
plt.figure(figsize=(10, 6))
top_5_series.plot(kind='bar')
plt.title('Top 5 TV Series with the Highest Number of Seasons')
plt.xlabel('TV Series')
plt.ylabel('Number of Seasons')
plt.xticks(rotation=35, ha='right')
plt.tight_layout()
plt.show()
```

```python
df['date_added'] = df['date_added'].fillna(df['date_added'].min())
df['year_added'] = df['date_added'].dt.year.astype(int)
```

### here we had to fill the empty values in date added withe the minimum date

```python
tv_shows = df[df['type'] == 'TV Show'].groupby('year_added').size()
movies = df[df['type'] == 'Movie'].groupby('year_added').size()
```

```python
plt.figure(figsize=(10, 6))
tv_shows.plot(label='Nouvelles sÃ©ries tÃ©lÃ©visÃ©es', marker='o')
movies.plot(label='Nouveaux films', marker='o')
plt.title('Nombre de Nouvelles SÃ©ries TÃ©lÃ©visÃ©es et de Nouveaux Films AjoutÃ©s par AnnÃ©e')
plt.ylabel('Nombre de Titres')
plt.legend()
plt.xticks(range(min(df['year_added']), max(df['year_added']) + 1))
plt.show()
```
